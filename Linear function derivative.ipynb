```python
import numpy as np
import matplotlib.pyplot as plt

# Define the linear function y = mx + c
m = 2  # slope
c = 3  # intercept
x = np.linspace(-10, 10, 100)
y = m * x + c

# Plot the function
plt.figure(figsize=(8, 5))
plt.plot(x, y, label=r'$y = 2x + 3$', color='blue')

# Plot the derivative (which is just the slope line)
plt.axhline(y=m, color='red', linestyle='--', label=r'Derivative: $m = 2$')

# Labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of a Linear Equation and its Derivative')
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
