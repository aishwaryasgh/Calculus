```python
import numpy as np
import matplotlib.pyplot as plt

# Define the constant function and its derivative
x = np.linspace(-10, 10, 400)
c = 1 # Example constant
f_x = np.full_like(x, c)  # Constant function f(x) = c
f_prime_x = np.zeros_like(x)  # Derivative is always 0

# Plot the function and its derivative
plt.figure(figsize=(8, 5))
plt.plot(x, f_x, label=r'$f(x) = c$', color='blue', linestyle='-')
plt.plot(x, f_prime_x, label=r"$f'(x) = 0$", color='red', linestyle='--')

# Labels and title
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graph of a Constant Function and Its Derivative")
plt.grid(True)

# Show the plot
plt.show()
```
